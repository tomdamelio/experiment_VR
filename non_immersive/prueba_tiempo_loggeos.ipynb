```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
df = pd.read_csv('./results/sub-45_ses-A_task-Experiment_VR_non_immersive_beh.csv')
```

```python
for i in [1,2,3,4]:
    for ii in ['continuous_annotation', 'continuous_annotation_luminance']:
        # Seleccionar los valores de la columna `continuous_annotation` para id == 1
        filtered_data = df.loc[df['id'] == i, ii].iloc[0]

        list_of_lists = eval(filtered_data)

        # Extraer solo los timestamps de cada sublista
        timestamps = [pair[1] for pair in list_of_lists]

        # Calcular la diferencia entre cada punto de dato (t) y el siguiente (t+1)
        differences = [timestamps[i+1] - timestamps[i] for i in range(len(timestamps) - 1)]

        # Crear un plot para visualizar las variaciones de tiempo
        plt.plot(differences)
        plt.xlabel('Index')
        plt.ylabel('Time Difference (s)')
        plt.title(f'{i} - {ii} - Time Differences Between Consecutive Timestamps')
        plt.show()
```
