```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast  # Utilizaremos ast.literal_eval para convertir el string a una lista de Python

# Lista de sujetos
subjects = ['sub-02', 'sub-03', 'sub-04', 'sub-05', 'sub-06']

# Configuracin de estilos para la presentacin
sns.set(style="whitegrid")
colors = sns.color_palette("tab10")

# Iterar sobre cada sujeto
for subject in subjects:
    # Cargar el archivo .csv correspondiente al sujeto
    data = pd.read_csv(f'../results/{subject}/ses-A/beh/{subject}_ses-A_task-Experiment_VR_non_immersive_beh.csv')

    # Filtrar filas con valor 'yes' en la columna 'luminance'
    filtered_data = data[data['luminance'] == 'yes']

    for stim in range(0, len(filtered_data)):
        video_id = filtered_data['id'].iloc[stim]

        # Convertimos el string a una lista de listas para continuous_annotation
        continuous_annotation = ast.literal_eval(filtered_data['continuous_annotation'].iloc[stim])
        y_values_ca = [pair[0] for pair in continuous_annotation]  # Primer valor de cada par como valor en y
        x_values_ca = [pair[1] for pair in continuous_annotation]  # Segundo valor de cada par como valor en x (tiempo)

        # Convertimos el string a una lista de listas para continuous_annotation_luminance
        continuous_annotation_luminance = ast.literal_eval(filtered_data['continuous_annotation_luminance'].iloc[stim])
        y_values_cal = [pair[0] for pair in continuous_annotation_luminance]  # Primer valor de cada par como valor en y
        x_values_cal = [pair[1] for pair in continuous_annotation_luminance]  # Segundo valor de cada par como valor en x (tiempo)

        # Creando el lineplot comparativo
        plt.figure(figsize=(16, 14))
        plt.plot(x_values_cal, y_values_cal, marker='o', linestyle='-', color=colors[1], label='True Luminance')
        plt.plot(x_values_ca, y_values_ca, marker='o', linestyle='-', color=colors[0], label='Annotated Luminance')
        
        plt.title(f'True vs. Annotated Luminance ({subject}) - ID: {video_id}', fontsize=44, fontweight='bold')
        plt.xlabel('Time (s)', fontsize=22)
        plt.ylabel('Luminance', fontsize=28)
        plt.ylim(-1.2, 1.2)  # Asegurando que el rango en y sea entre -1 y 1
        plt.legend(fontsize=30)
        plt.grid(True)
        
        # Mejorando la esttica del grfico
        plt.xticks(fontsize=24)
        plt.yticks(fontsize=24)
        plt.tight_layout()
        
        # Guardando la figura
        plt.savefig(f'../analysis/figures/luminance/comparison_plot_{video_id}_{subject}.png', dpi=300)
        plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast  # Utilizaremos ast.literal_eval para convertir el string a una lista de Python

# Carga del archivo .csv
data = pd.read_csv('../results/sub-06/ses-A/beh/sub-06_ses-A_task-Experiment_VR_non_immersive_beh.csv')

# Filtrar filas con valor 'yes' en la columna 'luminance'
filtered_data = data[data['luminance'] == 'yes']

for stim in range(0, len(filtered_data)):
    video_id = filtered_data['id'].iloc[stim]

    # Convertimos el string a una lista de listas
    first_value = ast.literal_eval(filtered_data['continuous_annotation'].iloc[stim])

    # Preparando los datos para el lineplot
    y_values = [pair[0] for pair in first_value]  # Primer valor de cada par como valor en y
    x_values = [pair[1] for pair in first_value]  # Segundo valor de cada par como valor en x (tiempo)

    # Creando el lineplot
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, marker='o', linestyle='-')
    plt.title(f'ID: {video_id} (continuous_annotation)')
    plt.xlabel('Time')
    plt.ylabel('Annotation Value')
    plt.ylim(-1.2, 1.2)  # Asegurando que el rango en y sea entre -1 y 1
    plt.show()

    # Convertimos el string a una lista de listas
    first_value = ast.literal_eval(filtered_data['stim_value'].iloc[stim])

    # Preparando los datos para el lineplot
    y_values = [pair[0] for pair in first_value]  # Primer valor de cada par como valor en y
    x_values = [pair[1] for pair in first_value]  # Segundo valor de cada par como valor en x (tiempo)

    # Creando el lineplot
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, marker='o', linestyle='-')
    plt.title(f'ID: {video_id} (stim_value)')
    plt.xlabel('Time')
    plt.ylabel('Annotation Value')
    plt.ylim(-1.2, 1.2)  # Asegurando que el rango en y sea entre -1 y 1
    plt.show()

    # Convertimos el string a una lista de listas
    first_value = ast.literal_eval(filtered_data['continuous_annotation_luminance'].iloc[stim])

    # Preparando los datos para el lineplot
    y_values = [pair[0] for pair in first_value]  # Primer valor de cada par como valor en y
    x_values = [pair[1] for pair in first_value]  # Segundo valor de cada par como valor en x (tiempo)

    # Creando el lineplot
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, marker='o', linestyle='-')
    plt.title(f'ID: {video_id} (continuous_annotation_luminance)')
    plt.xlabel('Time')
    plt.ylabel('Annotation Value')
    plt.ylim(-1.2, 1.2)  # Asegurando que el rango en y sea entre -1 y 1
    plt.show()
```
