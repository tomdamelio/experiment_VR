```python
# Primero, necesitaremos cargar el archivo .csv para poder inspeccionarlo y extraer los datos necesarios.
import pandas as pd
import matplotlib.pyplot as plt

# Carga del archivo .csv
file_path = "../results/sub-99/ses-A/beh/sub-99_ses-A_task-Experiment_VR_non_immersive_beh.csv"
data = pd.read_csv(file_path)
```

```python
import ast  # Utilizaremos ast.literal_eval para convertir el string a una lista de Python

for stim in range(0, len(data)):

    # Convertimos el string a una lista de listas
    first_value = ast.literal_eval(data['continuous_annotation'].iloc[stim])

    # Preparando los datos para el lineplot
    y_values = [pair[0] for pair in first_value]  # Primer valor de cada par como valor en y
    x_values = [pair[1] for pair in first_value]  # Segundo valor de cada par como valor en x (tiempo)

    # Creando el lineplot
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, marker='o', linestyle='-')
    plt.title(f'Affective Annotation Over Time - Stimulus {stim}')
    plt.xlabel('Time')
    plt.ylabel('Annotation Value')
    plt.ylim(-1.2, 1.2)  # Asegurando que el rango en y sea entre -1 y 1
    plt.show()

    # Convertimos el string a una lista de listas
    first_value = ast.literal_eval(data['stim_value'].iloc[stim])

    # Preparando los datos para el lineplot
    y_values = [pair[0] for pair in first_value]  # Primer valor de cada par como valor en y
    x_values = [pair[1] for pair in first_value]  # Segundo valor de cada par como valor en x (tiempo)

    # Creando el lineplot
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, marker='o', linestyle='-')
    plt.title(f'Luminance Over Time - Stimulus {stim}')
    plt.xlabel('Time')
    plt.ylabel('Annotation Value')
    plt.ylim(-1.2, 1.2)  # Asegurando que el rango en y sea entre -1 y 1
    plt.show()

    # Convertimos el string a una lista de listas
    first_value = ast.literal_eval(data['continuous_annotation_luminance'].iloc[stim])

    # Preparando los datos para el lineplot
    y_values = [pair[0] for pair in first_value]  # Primer valor de cada par como valor en y
    x_values = [pair[1] for pair in first_value]  # Segundo valor de cada par como valor en x (tiempo)

    # Creando el lineplot
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, marker='o', linestyle='-')
    plt.title(f'Luminance Annotation Over Time - Stimulus {stim}')
    plt.xlabel('Time')
    plt.ylabel('Annotation Value')
    plt.ylim(-1.2, 1.2)  # Asegurando que el rango en y sea entre -1 y 1
    plt.show()

```
